# Задание 1: Формирование дизайна A/B теста

В этом задании вам необходимо составить свой дизайн A/B-теста.

**Для этого вам нужно выбрать произвольный продукт, дать его краткое описание:**  
Возьмем наш выпускной проект MADE, над которым мы с командой сейчас работаем --- Toxicity-Checker Bot  
*Основная функция продукта:* Это телеграм-бот, который добавляется в группу, где общаются пользователи, и проверяет каждое сообщение в чате на токсичность. В случае, если какое-то сообщение превышает порог токсичности, оно классифицируется как токсичное и в чате бот информирует об этом своим сообщением.  
*Цель продукта:* сделать общение в чате комфортным и конструктивным для всех участников, снизить токсичность, путем "подсветки" для модераторов токсичных сообщений и их авторов.  


**Сформулировать дизайн A/B-теста, то есть определить следующие пункты:**  
Когда бот детектит токсичное сообщение, он сам, в свою очередь, оповещает об этом в чате сообщением. Кажется, что от формулировки данных сообщений может зависеть количество токсичных сообщений в данном чате, то есть может как вырасти, так и снизиться.   
Формулировка1(мягкая): "Сообщение \<message\> от пользователя \<username\> превышает допустимый уровень токсичности. Пользователь \<username\>, пожалуйста, используйте более вежливую и уважительную лексику, чтобы всем было комфортно и продуктивно общаться." --- объясняет "по-хорошему", для чего нужно общаться вежливо, а   
Формулировка2(жесткая): "Сообщение \<message\> от пользователя \<username\> превышает допустимый уровень токсичности. Пользователь \<username\>, используйте более вежливую и уважительную лексику, иначе Вы будете заблокированы и удалены из чата." --- объясняет через "силу", через наказание, которое последует за продолжением токсичного общения.
Проверим с помощью аб-теста какую формулировку лучше использовать
1. Определить изменение(гипотеза):  Использование мягкой формулировки1 вместо жесткой формулировки2 снижает уровень токсичности в чате
2. Определить метрики: number_toxic_messages / total_number_messages --- доля токсичных сообщений в чате за период тестирования
3. Определить аудиторию: Аудитория, в данном случае, это чаты в которые добавлен наш бот
4. Определить инфраструктуру: Разделим все чаты на две группы так, чтобы в каждой группе было одинаковое количество чатов с большим и маленьким кол-вом участников, с высокой и низкой ранее зафиксированной токсичностью, с высоким и низким количеством сообщений в день. Далее бот, перед отправкой сообщения, проверяет к какой группе принадлежит чат и отправляет соответствующее сообщение.
5. Определить длительность проведения теста: 2 месяца
6. Определить ожидаемый эффект: снижение доли токсичных сообщений на 10%
